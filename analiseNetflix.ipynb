```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./netflix_titles.csv')
```

```python
df.head(20)
```

```python
paises = df['country'].unique() #Recebe os países únicos da coluna de países 

mapQuantidades = {}

#Laço para criar um dicionário dinamicamente de acordo com a coluna country
for i in paises:
    
    mapQuantidades[i] = len(df['country'].where(df['country'] == i).dropna())

#Ordenação dos países por quantidade de filmes produzidos
ranking = dict( sorted(mapQuantidades.items(), key=lambda item: item[1], reverse=True))
        
#Exibe a tabela com os 50 primeiros no ranking
pd.DataFrame.from_dict(ranking, orient='index', columns = ['Filmes']).head(20)
```

```python

#String única para receber todas as categorias existentes no csv
categoriasRaw = ""

#Montagem da string com categorias encontradas
for i in df['listed_in'].unique():
    categoriasRaw += i + ","
    
#Divide as categorias quebradas por vírgula e retornando um array
categoriasSplit = categoriasRaw.split(',')

#Retirar todos os espaços ao redor de cada elemento (categoria)
for i in range(len(categoriasSplit)):
    categoriasSplit[i] = categoriasSplit[i].strip()

#Traz as categorias únicas, removendo repetições
categorias = np.unique(categoriasSplit)

#Remove o índice 0 que possui um espaço em branco apenas
categorias = np.delete(categorias, 0)

categorias
```

```python

# Ler linha a linha do csv

# Criar um dict para armazenar os países

dictPaises = {}

for i in paises:
    dictPaises[i] = {}
    for j in categorias:
        if(j == 'Movies'):
            continue
        lista = df.where(df['country'] == i).dropna()
        cat = lista['listed_in']
        dictPaises[i][j] = 0;
        for c in cat:
            if(c.find(j) != -1):                   
                dictPaises[i][j] += 1        
```

```python
for i in paises:
    dictPaises[i] = dict( sorted(dictPaises[i].items(), key=lambda item: item[1], reverse=True))
    
# print(dictPaises)    
```

```python
# Fazer novo dict com as 3 primeiras categorias de cada país

rankingCategorias = {}

for i in paises:
    
    rankingCategorias[i] = dict(list(dictPaises[i].items())[:3])
    
    
# print(rankingCategorias)
```

```python
# Criar tabela com os dados do novo dict

pd.DataFrame.from_dict({(i,j): rankingCategorias[i][j] 
                           for i in rankingCategorias.keys() 
                           for j in rankingCategorias[i].keys()},
                       orient='index', columns = ['Quantidades']).head(25)
```
